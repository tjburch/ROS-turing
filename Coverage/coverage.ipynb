# Regression and Other Stories: Coverage

Coverage - Illustration of coverage of intervals. See Chapter 4 in Regression and Other Stories.

-----------

## Simulate

```julia
using Distributions, StatsPlots
```

```julia
n_rep = 100
est = fill(NaN, n_rep)
conf = fill(NaN, (n_rep, 4))
mu = 6
sigma = 4

for i in 1:100
    y = rand(Normal(mu, sigma))
    est[i] = y
    conf[i,:] = y .+ [-2, -.67, .67, 2] * sigma
end
```

## Plot

```julia
x = 1:n_rep
y = est
sig1= (conf[:,2])#; conf[:,3])
scatter(
    x,
    y,
    yerr=conf[:,3] - conf[:,2],
    legend=false,
    ylabel="Estimate, 50%, and 95% confidence interval",
    xlabel="Simulation"
)
scatter!(
    x,
    y,
    yerr=conf[:,4] - conf[:,1] # TODO - find a less ugly way to plot this
)
hline!(fill(0,100), line=(:gray40))
```
