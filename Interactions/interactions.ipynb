# Regression and Other Stories: Interaction example

Plot interaction example figure. See Chapter 1 in Regression and Other Stories.

-----

### Load packages

```julia
using Plots
```

### Plot interaction example figure

```julia
plot(
    0:20, x->0.07409+x .*0.0134,
    color=:black,
    legend=false,grid=false,
    xlabel="Home radon exposure (pCi/L)",
    ylabel="Probability of lung cancer",
    title="Example of an interaction",
    xlim=(0,12.5), ylim=(0,0.25)
)
plot!(0:20, x->0.00579+x .*0.0026, color=:black)
annotate!(10, .07409+10*.0134 - .02, text("Smokers", 9))
annotate!(10, .00579+10*.0026 + .01, text("Nonsmokers", 9))
```
