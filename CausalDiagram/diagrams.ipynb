# Regression and Other Stories: CausalDiagram

Plot a causal diagram. See Chapter 19 in Regression and Other Stories.

----

#### Load packages

```julia
using Plots, LaTeXStrings
```

#### Plot causal diagram

```julia
plot([1.1,3.3],[1.1,0.5],arrow=true,color=:black,legend=false,
xlim=(-0.5,4),ylim=(0,2), grid=false, xaxis=false, yaxis=false, xticks=false, yticks=false)
plot!([1.1,3.3],[1.1,1.5], arrow=true, color=:black, )
annotate!(0, 1, "x", :bottom)
annotate!(1, 1, "z",:bottom)
annotate!(3.5, .5, LaTeXString("\$y^0\$"), 14)
annotate!(3.5, 1.5, LaTeXString("\$y^1\$"), 14)
annotate!([0,1,3.5], [1.8,1.8,1.8], ["Before","Treatment","Outcome"], 10)
annotate!(1.9, .72, text("Treatment 0 (z=0)", rotation=-16,:bottom))
annotate!(1.9, 1.27, text("Treatment 1 (z=1)", rotation=14.3, :bottom))
annotate!(1.8, .1, "Time")
plot!([-.3,4.0],[.2,0.2], arrow=true, color=:black)

```
