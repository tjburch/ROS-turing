# Regression and Other Stories: Metabolic

An example of how to interpret a power law or log-log regression. See Chapter 3 in Regression and Other Stories.

Animals' body mass and metabolism comes from section 3.8.2 of Gelman and Nolan: Teaching Statistics: A Bag of Tricks, second edition. Oxford University Press, 2017.

----

## Load packages

```julia
using StatsPlots, LaTeXStrings
```

```julia
labels = ["Mouse", "Man", "Elephant"]
x = [log(0.02), log(65), log(3000)]
y = [log(0.17), log(90), log(2000)]
scatter(
    x, y,
    xlabel="log (body mass in kilograms)",
    ylabel="log (metabolic rate in watts)",
    legend=false
)

plot!(
    log(0.01):log(10000),
    x-> 1.4 + 0.74 * x,
    color=:black
)
annotate!(x[1] + 0.3,y[1], text(labels[1], 8, :left))
annotate!(x[2] - 0.3,y[2], text(labels[2], 8, :right))
annotate!(x[3] - 0.3,y[3], text(labels[3], 8, :right))
annotate!(log(2), log(3.5), text("log y = 1.4 + 0.74 log x", 8, :left))
```

```julia
labels = ["Mouse", "Man", "Elephant"]
x = [0.02, 65, 3000]
y = [0.17, 90, 2000]
scatter(
    x, y,
    xlabel="log (body mass in kilograms)",
    ylabel="log (metabolic rate in watts)",
    legend=false
)

plot!(
    0:3000,
    x-> exp(1.4 + 0.74 * log(x)),
    color=:black
)
annotate!(x[1] + 50,y[1], text(labels[1], 8, :left))
annotate!(x[2] -10 ,y[2] + 35, text(labels[2], 8, :bottom))
annotate!(x[3] - 40,y[3], text(labels[3], 8, :right))
annotate!(
    1250, 700, 
    text(
        LaTeXString("\$y = e^{1.4 + 0.74 \\log x} = 4.1 x^{0.74}\$"), 
        10, :left
    )
)
```
