# Regression and Other Stories: Balance

Figures for hypothetical example of zero causal effect but positive predictive comparison. See Chapter 20 in Regression and Other Stories.

----

## Load packages

```julia
using Distributions, StatsPlots
```

## Plot figures

```julia
function bell(filename, mu, sd, lo, hi, ymax)
    p = plot(
        Normal(mu,sd),
        color=:black,
        xlabel="", ylabel="", title="",
        yaxis=false, xticks=1:5,yticks=false,
        ylim=(0,ymax), xlim=(lo,hi), legend=false
    )
    savefig(filename)
    display(p)
end;
```
