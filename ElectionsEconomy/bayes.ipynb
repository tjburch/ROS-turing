# Regression and Other Stories: Elections Economy - Bayes

Demonstration of Bayesian information aggregation. See Chapter 9 in Regression and Other Stories.

----

## Load packages

```julia
using StatsPlots, Distributions
```

## Calculations

Prior based on a previously-fitted model using economic and political condition.

```julia
theta_hat_prior = 0.524
se_prior = 0.041;
```

Survey of 400 people, of whom 190 say they will vote for the Democratic candidate

```julia
n = 400;
y = 190;
```

## Data estimate

```julia
theta_hat_data = y/n;
se_data = sqrt((y/n)*(1-y/n)/n);
```

## Bayes estimate

```julia
theta_hat_bayes = (theta_hat_prior/se_prior^2 + theta_hat_data/se_data^2) / 
(1/se_prior^2 + 1/se_data^2)
se_bayes = sqrt(1/(1/se_prior^2 + 1/se_data^2));
```

## Figures

```julia
plot(Normal(theta_hat_prior, se_prior),
    color=:black, ylim=(0,NaN),
    legend=false,yaxis=false,yticks=false
)
annotate!(0.588, 5, text("Prior", 10))
plot!(Normal(theta_hat_data, se_data), color=:black)
annotate!(0.420, 8, text("Likelihood", 10))
```

```julia
plot(Normal(theta_hat_prior, se_prior),
    color=:black, ylim=(0,NaN),
    legend=false,yaxis=false,yticks=false
)
annotate!(0.588, 5, text("Prior", 10))
plot!(Normal(theta_hat_data, se_data), color=:black)
annotate!(0.420, 8, text("Likelihood", 10))
plot!(Normal(theta_hat_bayes, se_bayes), color=:black)
annotate!(0.525, 15, text("Posterior", 10))
```
